### Анализ и обработка пропусков и приведение данных к нужным типам

```python
# проверка наличия пропусков по столбцам по количеству
display(data.isna().sum())
display(data.isnull().sum())

# заполним все пропуски как unknown или нулевыми значениями
data['name'] = data['name'].fillna('unknown')
data['genre'] = data['genre'].fillna('unknown')

# замена значений tbd на NaN
for i in range(len(data)):
    if data.loc[i, 'user_score'] == 'tbd':
        data.loc[i, 'user_score'] = 'NaN'

# округление до десятых значений столбцов na_sales, eu_sales, jp_sales, other_sales
for i in range(len(data)):
    data.loc[i, 'na_sales'] = round(data.loc[i, 'na_sales'], 1)
    data.loc[i, 'eu_sales'] = round(data.loc[i, 'eu_sales'], 1)    
    data.loc[i, 'jp_sales'] = round(data.loc[i, 'jp_sales'], 1)    
    data.loc[i, 'other_sales'] = round(data.loc[i, 'other_sales'], 1)
    
# приведем типы данных существующей таблицы к более подходящим методом astype
try:
    data['year_of_release'] = data['year_of_release'].astype('Int64')
    data['critic_score'] = data['critic_score'].astype('Int64')    
    data['user_score'] = data['user_score'].astype('float64')        
except:
    print('В данных присутствует другая необработанная ошибка, необходимо проанализировать данные поподробнее')
    
# проверка датасета
display(data.info())
display(data.head(10))
```

```python
# посмотрим как влияют отзывы пользователей и критиков на продажи внутри популярной платформы PS4 
# приведем значения столбца critic_score к масштабу значений столбца user_score поделив значения critic_score на 10
data_short_cycle['critic_score'] = data_short_cycle['critic_score'] / 10

# сформируем таблицу для анализа платформы
data_short_cycle_ps4 = data_short_cycle.query('platform == "PS4"').reset_index(drop=True)
data_short_cycle_ps4_critic = data_short_cycle_ps4.dropna(subset=['critic_score'])
data_short_cycle_ps4_user = data_short_cycle_ps4.dropna(subset=['user_score'])

# построим диаграммы рассеяния для представления корреляции между оценками критиков и пользователей и объемов продаж
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15,5))

# влияние на продажи отзывов критиков
ax1.scatter(x=data_short_cycle_ps4_critic['critic_score'], y=data_short_cycle_ps4_critic['summary_sales'])
ax1.set_title('Влияние на продажи отзывов критиков')
ax1.set_xlabel('Оценки критиков')
ax1.set_ylabel('Объем продаж в млн. копий')

# влияние на продажи отзывов пользователей
ax2.scatter(x=data_short_cycle_ps4_user['user_score'], y=data_short_cycle_ps4_user['summary_sales'])
ax2.set_title('Влияние на продажи отзывов пользователей')
ax2.set_xlabel('Оценки пользователей')
ax2.set_ylabel('Объем продаж в млн. копий')
plt.show()

# расчет коэффициента корреляции для оценок критиков и для оценок пользователей
print('Коэффициент корреляции Пирсона по объему продаж и отзывам критиков: ', data_short_cycle_ps4['summary_sales'].corr(data_short_cycle_ps4['critic_score']))
print('Коэффициент корреляции Пирсона по объему продаж и отзывам пользователей: ', data_short_cycle_ps4['summary_sales'].corr(data_short_cycle_ps4['user_score']))
```

## Проверка гипотез

### Гипотезы для проверки:
1. ГИПОТЕЗА №1: средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Примем следующую гипотезу за нулевую - средние пользовательские рейтинги платформы Xbox One равны средним пользовательским рейтингам платформы PC. Исходя из нулевой гипотезы, примем за альтернативную гипотезу - средние пользовательские рейтинги платформ не равны.
2. ГИПОТЕЗА №2: средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные. Примем следующую гипотезу за нулевую - средние пользовательские рейтинги жанра Action равны средним пользовательским рейтингам жанра Sports. Исходя из нулевой гипотезы, примем за альтернативную гипотезу - средние пользовательские рейтинги жанров не равны.

Примечание: пороговое значение alpha задано как 0.05. Т.е. принятый уровень значимости для нас более 5%.

Критерии для проверки: метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: scipy.stats.ttest_ind используется для проверки Гипотезы №1, №2: выборки независимы, для корректного использования t-теста необязательно, чтобы распределение было нормальным(см. примечание ниже), дисперсии не равны(см. расчет ниже), но этот метод выбирается из расчета на то, что, как правило, этот тест редко ошибается.

Примечание: вне зависимости от того, распределена ли исходная величина нормально или нет, на практике мы можем считать, что выборочное среднее распределено нормально и мы можем применять t-тест, если размер выборки достаточно большой - скажем, больше 50. Или 100. Или 500.

```python
# тест ГИПОТЕЗЫ №1
# проверки гипотезы о равенстве среднего двух генеральных совокупностей(пользовательские рейтинги платформ Xbox One и PC) по имеющимся выборкам
array_xone = data_short_cycle_xone_user.query('user_score > 0')['user_score']
array_pc = data_short_cycle_pc_user.query('user_score > 0')['user_score']

# вывод аналитической информации
print('Дисперсия выборки рейтинга платформы Xbox One: ', np.var(array_xone))
print('Дисперсия выборки рейтинга платформы PC: ', np.var(array_pc))
print('Среднее выборки рейтинга платформы Xbox One: ', np.mean(array_xone))
print('Среднее выборки рейтинга платформы PC: ', np.mean(array_pc))
print('Стандартное отклонение выборки рейтинга платформы Xbox One: ', np.std(array_xone))
print('Стандартное отклонение выборки рейтинга платформы PC: ', np.std(array_pc))
print('Величина выборки рейтинга платформы Xbox One: ', len(array_xone))
print('Величина выборки рейтинга платформы PC: ', len(array_pc))

# задан уровень статистической значимости 5%
alpha = 0.05

print('Уровень статистической значимости {:.1%}'.format(alpha))

# тест ГИПОТЕЗЫ №1 при equal_var=False(считаем не равными дисперсии выборок)
results_first_test = st.ttest_ind(array_xone, array_pc, equal_var=False)

# тест двусторонний
print('p-значение в %:', results_first_test.pvalue * 100)

# отвергаем гипотезу только тогда, когда выборочное среднее значимо меньше предполагаемого значения
if (results_first_test.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не отвергаем нулевую гипотезу") 
```

Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно почти 28.67036680187632%. Это достаточно невысокая вероятность, чтобы делать вывод о значимом различии между пользовательскими рейтингами платформы Xbox One и PC. При этом был использован equal_var = False, т.к. выборки все же отличаются: по Xbox One - 116 строки датасета, по PC - 110 строки датасета, а дисперсии по PC - 3, по Xbox One - 1.6. Результаты теста говорят о том, что статистически значимая разница между двумя средними по Xbox One и PC выше среднего, однако обозначенную ранее нулевую гипотезу равенства средних двух выборок по рейтингам пользователей Xbox One и PC нельзя отвергать, что, соответственно, не подтверждает альтернативную гипотезу о различии средних выборок по пользовательским рейтингам этих платформ, что подтверждает сформулированную в задаче Гипотезу №1 "средние пользовательские рейтинги платформ Xbox One и PC одинаковые" - принимаем к рассмотрению эту гипотезу.

```python
# тест ГИПОТЕЗЫ №2
# проверки гипотезы о равенстве среднего двух генеральных совокупностей(пользовательские рейтинги жанров Action и Sports) по имеющимся выборкам
array_action = data_short_cycle.query('(user_score > 0) & (genre == "Action")')['user_score']
array_sports = data_short_cycle.query('(user_score > 0) & (genre == "Sports")')['user_score']

# вывод аналитической информации
print('Дисперсия выборки рейтинга жанра Action: ', np.var(array_action))
print('Дисперсия выборки рейтинга жанра Sports: ', np.var(array_sports))
print('Среднее выборки рейтинга жанра Action: ', np.mean(array_action))
print('Среднее выборки рейтинга жанра Sports: ', np.mean(array_sports))
print('Стандартное отклонение выборки рейтинга жанра Action: ', np.std(array_action))
print('Стандартное отклонение выборки рейтинга жанра Sports: ', np.std(array_sports))
print('Величина выборки рейтинга жанра Action: ', len(array_action))
print('Величина выборки рейтинга жанра Sports: ', len(array_sports))

# задан уровень статистической значимости 5%
alpha = 0.05

print('Уровень статистической значимости {:.1%}'.format(alpha))

# тест ГИПОТЕЗЫ №2 при equal_var=False(считаем не равными дисперсии выборок)
results_second_test = st.ttest_ind(array_action, array_sports, equal_var=False)

# тест двусторонний
print('p-значение в %:', results_second_test.pvalue * 100)

# отвергаем гипотезу только тогда, когда выборочное среднее значимо меньше предполагаемого значения
if (results_second_test.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не отвергаем нулевую гипотезу") 
```

Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно почти 3.507673706253368e-17%, т.е. стремится к 0. Вероятность слишком мала. При этом был использован equal_var = False, т.к. выборки по размерам значительно отличаются: по жанру Action - 304 строки датасета, по жанру Sports - 123 строки датасета, дисперсия по данным выборкам также значительно отличается. Результаты теста говорят о том, что статистически значимая разница между двумя средними жанра Action и жанра Sports существует, следовательно обозначенную ранее нулевую гипотезу равенства средних двух выборок по жанру Action и по жанру Sports можно отвергнуть, что, соответственно, может подтвердить альтернативную гипотезу о различии средних выборок жанрам, что не опровергает сформулированную в задаче Гипотезу №2 "средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные" - оставляем эту гипотезу.

Были проверены следующие гипотезы:
   - ГИПОТЕЗА №1: средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Данная гипотеза подтверждена, большая вероятность, чтобы делать вывод о значимом различии между пользовательскими рейтингами платформы Xbox One и PC. Результаты теста говорят о том, что статистически значимая разница между двумя средними по Xbox One и PC мала, следовательно гипотезу равенства средних двух выборок по рейтингам пользователей Xbox One и PC нельзя отвергать, что, соответственно, не подтверждает альтернативную гипотезу о различии средних выборок по пользовательским рейтингам этих платформ, что подтверждает сформулированную в задаче Гипотезу №1, которую мы принимаем к рассмотрению;
   - ГИПОТЕЗА №2: средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные. Данная гипотеза подтверждена, результаты теста говорят о том, что статистически значимая разница между двумя средними жанра Action и жанра Sports существует, поэтому оставляем эту гипотезу.

В рамках исследования были сформулированы следующие рекомендации по корректировке маркетинговых кампаний для максимизации продаж и прибыли:
1. необходимо проанализировать текущее состояние современного рынка игр, наложить данные на анализ за актуальный период и выявить платформы, на которые необходимо сделать акцент при продаже;
2. сделать акцент на отзывы критиков, которые имеют наибольшее влияние на объемы продаж, провести дополнительный анализ и постараться выявить дополнительные факторы, которые могут иметь большую корреляцию с продажами игр пользователям, т.е. уточнить пользовательский опыт;
3. оптимизировать маркетинговые кампании в регионах присутствия в зависимости от пользовательского опыта конкретного региона выявленного на этапе исследования и в зависимости от конверсии по каждому жанру;
4. учесть особенности платформ в рекламных кампаниях по регионам(Япония, например, предпочитает более локальные истории по продуктам);
5. учесть влияние рейтинга ESRB на структуру спроса, в том числе по регионам, а также улучшить качество входной информации для анализа и проведения исследований.
