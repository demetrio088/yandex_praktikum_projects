# Исследование объявлений о продаже квартир

В нашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Необходимо определить рыночную стоимость объектов недвижимости. Задача — установить параметры. ### Особенности и обзор данных

Согласно документации к данным выделены следующие столбцы/сущности в наборах данных:
* airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
* balcony — число балконов
* ceiling_height — высота потолков (м)
* cityCenters_nearest — расстояние до центра города (м)
* days_exposition — сколько дней было размещено объявление (от публикации до снятия)
* first_day_exposition — дата публикации
* floor — этаж
* floors_total — всего этажей в доме
* is_apartment — апартаменты (булев тип)
* kitchen_area — площадь кухни в квадратных метрах (м²)
* last_price — цена на момент снятия с публикации
* living_area — жилая площадь в квадратных метрах (м²)
* locality_name — название населённого пункта
* open_plan — свободная планировка (булев тип)
* parks_around3000 — число парков в радиусе 3 км
* parks_nearest — расстояние до ближайшего парка (м)
* ponds_around3000 — число водоёмов в радиусе 3 км
* ponds_nearest — расстояние до ближайшего водоёма (м)
* rooms — число комнат
* studio — квартира-студия (булев тип)
* total_area — площадь квартиры в квадратных метрах (м²)
* total_images — число фотографий квартиры в объявлении

В выведенной таблице на основе предоставленного массива данных обнаружены следующие особенности на этапе экспресс-анализа:
1. в столбцах last_price и total_area отражены значения типа float64, их для удобства отображения трансформируем в int64, т.к. погрешность при таком изменении минимальна, а изучать данные будет удобнее и нагляднее;
2. около половины строк данных по сущности ceiling_height пропущено, информация о высоте потолков должна быть вероятнее всего характерной для разных типов постройки, серий домов, но т.к. этой информации нет, то следует заполнить пропуски медианными значениями по выборке;
3. пропуски в floors_total лучше отсечь, т.к. менее 100 позиций по пропущенным строкам не дает погрешность более 1% от общей величины выборки, кроме того формат float64 целесообразно привести к int64, т.к. этажность всегда отражается в целых числах;
4. в столбцах total_area, living_area и kitchen_area для удобства отображения трансформируем в int64, т.к. погрешность при таком изменении минимальна, а изучать данные будет удобнее и нагляднее, также имеющиеся пропуски в полях(около 2000 строк по каждому из столбцов) можно восстановить по разнице между total_area и ненулевыми данными ячейки одного из столбцов living_area и kitchen_area, при этом значения kitchen_area заполним медианными, влияние данного показателя сгладит выборку по причине относительно малого количества данных незаполненных строк, но не исказит ее чрезмерно;
5. в столбце is_apartment значительное количество пропусков, незаполненные строки целесообразней заполнить 0, т.к. доля апартаментов по статистике значительно меньше жилого фонда(при условии, что для России это относительно новый вид бизнеса - продажа апартаментов), также следует перевести данные в формат bool согласно документации;
6. число заполненных строк в столбце balcony около половины, отсутствующие данные следует заполнить 0 значениями ввиду сложности моделирования количества балконов под разные типы планировок и районов Питера, кроме того следует перевести данные в формат int64 с точки зрения того что дробного количества балконов невозможно представить в реальности;
7. количество пропусков в locality_name незначительно, следует переименовать строки с неизвестными локациями на "unknown", данные неизвестны и потребуются дополнительные уточнения;
8. данные столбца airports_nearest указаны в метрах, следует перевести в километры для удобства восприятия данных и перевести в целочисленный формат int64. Отсутствующие данные следует заполнить медианным расстоянием до аэропорта/ов района принадлежности объекта, отсутствующие данные заполняются 0 и будут при необходимости отсечены, либо уточнены при дополнительном запросе в службу, которая предоставляла данные, если это возможно;
9. столбец cityCenters_nearest во первых следует переименовать в city_centers_nearest, затем следует перевести в километры для удобства восприятия данных и перевести в целочисленный формат int64. Отсутствующие данные(около 5000 пропусков) следует заполнить медианным расстоянием до аэропорта/ов района принадлежности объекта, отсутствующие данные заполняются 0 и будут при необходимости отсечены, либо уточнены при дополнительном запросе в службу, которая предоставляла данные, если это возможно;
10. в столбце parks_around3000 около 5000 пропусков, но данная информация зависит от удаленности от ближайших парков и от района объекта, заполнение оставшихся пропусков следует произвести по среднему количеству парков района принадлежности каждого конкретного жилья, отсутствующие данные заполняются 0 и будут при необходимости отсечены, либо уточнены при дополнительном запросе в службу, которая предоставляла данные, если это возможно. ### Обработка пропусков

```python
# подсчет пропусков
display(data.isna().sum()) # NaN
display(data.isnull().sum()) # '0'
```

```python
# заполняем данные по высоте потолков locality_name по выборке
ceiling_height_median = data[data['ceiling_height'] > 0]['ceiling_height'].median()
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_height_median)

# удаляем значения по публикациям с отсутствующей информацией по этажу
data.dropna(subset=['floors_total'], inplace=True)
data = data.reset_index(drop=True)

# в столбцах living_area и kitchen_area заполним пропуски разницей между значениями столбцов total_area и living_area/kitchen_area, остаток проанализируем и отсечем
kitchen_area_median = data[data['kitchen_area'] > 0]['kitchen_area'].median() # расчет медианной площади кухни для заполнения пропусков

for i in range(len(data)):
    if (pd.isna(data.loc[i, 'living_area']) == True or pd.isnull(data.loc[i, 'living_area']) == True):
        data.loc[i, 'living_area'] = data.loc[i, 'total_area'] - data.loc[i, 'kitchen_area']
    if (pd.isna(data.loc[i, 'kitchen_area']) == True or pd.isnull(data.loc[i, 'kitchen_area']) == True):
        data.loc[i, 'kitchen_area'] = data.loc[i, 'total_area'] - data.loc[i, 'living_area']

for i in range(len(data)):
    if (pd.isna(data.loc[i, 'living_area']) == True or pd.isnull(data.loc[i, 'living_area']) == True):
        data.loc[i, 'living_area'] = data.loc[i, 'total_area'] - kitchen_area_median
    if (pd.isna(data.loc[i, 'kitchen_area']) == True or pd.isnull(data.loc[i, 'kitchen_area']) == True):
        data.loc[i, 'kitchen_area'] = kitchen_area_median       

# заполняем 0 значениями столбец is_apartment        
data['is_apartment'] = data['is_apartment'].fillna(0)

# заполняем 0 значениями столбец balcony        
data['balcony'] = data['balcony'].fillna(0)

# заполняем "unknown" значениями столбец locality_name        
data['locality_name'] = data['locality_name'].fillna('unknown')

# заполним медианным расстоянием расстояние до аэропорта/ов, отсутствующие данные заполним 0 значениями
data['airports_nearest'] = data['airports_nearest'].fillna(data.groupby('locality_name')['airports_nearest'].transform('median'))     
data['airports_nearest'] = data['airports_nearest'].fillna(0)

# заполним медианным расстоянием расстояние до городского центра, отсутствующие данные заполним 0 значениями
data['city_centers_nearest'] = data['city_centers_nearest'].fillna(data.groupby('locality_name')['city_centers_nearest'].transform('median'))     
data['city_centers_nearest'] = data['city_centers_nearest'].fillna(0)     
        
# заполним средним количеством парков данные по паркам в округе, отсутствующие данные заполним 0 значениями
data['parks_around_3_km'] = data['parks_around_3_km'].fillna(data.groupby('locality_name')['parks_around_3_km'].transform('mean'))     
data['parks_around_3_km'] = data['parks_around_3_km'].fillna(0)     
        
# заполним медианным расстоянием расстояние до парков, отсутствующие данные заполним 0 значениями
data['parks_nearest'] = data['parks_nearest'].fillna(data.groupby('locality_name')['parks_nearest'].transform('median'))     
data['parks_nearest'] = data['parks_nearest'].fillna(0) 

# заполним средним количеством водоемов данные по водоемам в округе, отсутствующие данные заполним 0 значениями
data['ponds_around_3_km'] = data['ponds_around_3_km'].fillna(data.groupby('locality_name')['ponds_around_3_km'].transform('mean'))     
data['ponds_around_3_km'] = data['ponds_around_3_km'].fillna(0)     

# заполним медианным расстоянием расстояние до водоемов, отсутствующие данные заполним 0 значениями
data['ponds_nearest'] = data['ponds_nearest'].fillna(data.groupby('locality_name')['ponds_nearest'].transform('median'))     
data['ponds_nearest'] = data['ponds_nearest'].fillna(0)    
```

### Изучение квартир, выявление центральной области и расчет средней цены километра. Построение графиков

```python
# выделим квартиры в Санкт-Петербурге и добавим столбцы по средней стоимости 1 км. и переведем расстояние до центра в километры из метров
data['city_centers_nearest_km'] = (data['city_centers_nearest'] / 1000).astype('int64')
st_peterburg_data = pd.DataFrame(data.query('locality_name == "Санкт-Петербург"'))

# добавление категории расстояния: 1, 2..10
def km_category(km_data):
    km_total = km_data['city_centers_nearest_km']
    if km_total <= 1:
        return 1
    if (km_total <= 2) & (km_total > 1):
        return 2
    if (km_total <= 3) & (km_total > 2):
        return 3    
    if (km_total <= 4) & (km_total > 3):
        return 4    
    if (km_total <= 5) & (km_total > 4):
        return 5    
    if (km_total <= 6) & (km_total > 5):
        return 6   
    if (km_total <= 7) & (km_total > 6):
        return 7    
    if (km_total <= 8) & (km_total > 7):
        return 8    
    if (km_total <= 9) & (km_total > 8):
        return 9     
    if (km_total <= 10) & (km_total > 9):
        return 10
    return 0  
st_peterburg_data['km_to_center'] = st_peterburg_data.apply(km_category, axis=1)
st_peterburg_data['km_to_center'] = st_peterburg_data['km_to_center'].astype('int64')
# создание сводной таблицы по км до центра
pivot_nearest = st_peterburg_data.query('km_to_center > 0').pivot_table(index='km_to_center', 
                                          values='square_meter_cost', 
                                          aggfunc='mean', 
                                          margins=True)
display(pivot_nearest)

# вывод графика #2
pivot_nearest.plot()
plt.show()
```

Cтоимость 1 км. меняется от резкого снижения к росту в 6 км. от центра, затем цены снижаются.
