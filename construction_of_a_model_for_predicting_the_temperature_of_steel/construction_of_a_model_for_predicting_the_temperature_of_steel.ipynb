#### Описание процесса производства

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. 

Этапы обработки:
1. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша;
2. Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы; 
3. Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ;
4. Температуру на несколько минут повышают;
5. Сталь легируют, т.е. изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»);
6. Продувают сплав инертным газом;
7. Сплав перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки;
8. Расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки; 
9. Готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»).

В данной задаче предстоит разработать предсказательную модель температуры расплавленного состояния сплава, которая поспособствует бизнесу в работе над своими издержками.

```python
# обзор данных о подаче сыпучих материалов (время)
display(data_bulk_time.head(10))
display(data_bulk_time.tail(10))
display(data_bulk_time.info())
display(data_bulk_time.describe())
display(data_bulk_time.columns)
```

```python
# проанализируем количество замеров и распределение: количество замеров варьируется от 1(данные исключены) до 16
aggfuncs = ['first', 'last', 'count']
pivot_table_temp_count = pd.DataFrame(data_temp.groupby('key')['Время замера'].agg(aggfuncs))
pivot_table_temp_count = pivot_table_temp_count.query('count > 1')
display(pivot_table_temp_count.head(10))
display(pivot_table_temp_count.info())
pivot_table_temp_count['count'].hist(bins=20)
plt.title('Распределение количества замеров')
plt.show()
```

```python
# уберем партии, в которых последний срок замера раньше одного из добавлений сыпучих или проволочных материалов
pivot_table_temp_count = pivot_table_temp_count.merge(data_bulk_time, left_on='key', right_on='key', how='left')
pivot_table_temp_count = pivot_table_temp_count.merge(data_wire_time, left_on='key', right_on='key', how='left')
pivot_table_temp_count = pivot_table_temp_count.drop('count', axis=1)
pivot_table_temp_count.head(10)
```

```python
# построчно проверим являются ли последние замеры по каждой из партий действительно последними замерами,
# отбросим несоответствующие данные
pivot_table_temp_count['max'] = pivot_table_temp_count.max(axis=1)
pivot_table_temp_count.head(10)
pivot_table_temp_count.info()
```

```python
# если время последнего замера более позднее, то оставляем запись, если нет, то отбрасываем эту партию
pivot_table_temp_count = pivot_table_temp_count.query('last >= max')
display(pivot_table_temp_count.info())
```

```python
# сведем крайние температурные значения в таблицу по ключам для дальнейшего использования
pivot_table_last_temp_count = pivot_table_temp_count.merge(data_temp, left_on='last', right_on='Время замера', how='left')
pivot_table_last_temp_count = pivot_table_last_temp_count.drop('last', axis=1)
pivot_table_last_temp_count.set_index('key')
pivot_table_last_temp_count = pivot_table_last_temp_count.pivot_table(index='key', values='Температура')
display(pivot_table_last_temp_count.info())
display(pivot_table_last_temp_count.head(10))
display(pivot_table_last_temp_count.describe())
```

```python
# разделим выборки на целевую и основную
target = pivot_table_final['last_temperature']
features = pivot_table_final.drop('last_temperature', axis=1)
display(target.head(10))
display(target.shape)
display(features.head(10))
display(features.shape)
```

```python
# разобьем выборку на обучающую и валидационную для подбора лучших параметров моделей
features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=0.25, random_state=654321)
display(features_train.shape)
display(features_valid.shape)
display(target_train.shape)
display(target_valid.shape)
```

В данном разделе признаки модели были масштабированы с целью исключения наибольшего влияния признаков с более крупным значениями. Выборки были разделены на обучающую и валидационную.

```python
# анализ значимости факторов модели при помощи SHAP
shap.initjs()
explainer = shap.Explainer(rfr)
shap_test = explainer(features_valid)
shap_df = pd.DataFrame(shap_test.values, columns=shap_test.feature_names, index=features_valid.index)
columns = shap_df.apply(np.abs).mean().sort_values(ascending=False).index
fig, ax = plt.subplots(1, 2, figsize=(15,7))
sns.barplot(data=shap_df[columns].apply(np.abs), orient='h', ax=ax[0])
ax[0].set_title("Среднее абсолютное значение SHAP")
sns.boxplot(data=shap_df[columns], orient='h', ax=ax[1])
ax[1].set_title("Распределение значений SHAP")
plt.show()
```
