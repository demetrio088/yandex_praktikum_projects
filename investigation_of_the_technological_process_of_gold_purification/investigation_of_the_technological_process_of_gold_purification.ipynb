### Проверка расчета эффективности обогащения. Рассчет эффективности и MAE.

На данном этапе проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака rougher.output.recovery. Найдем MAE между нашими расчётами и значением признака. Сделаем выводы.

Расчет эффективности обогащения выглядит следующим образом:


где:

    C — доля золота в концентрате после флотации/очистки;
    F — доля золота в сырье/концентрате до флотации/очистки;
    T — доля золота в отвальных хвостах после флотации/очистки.

```python
# расчет эффективности обогащения
def recovery(c, f, t):
    result = (c * (f - t) / (f * (c - t))) * 100
    return result

data_train['rougher.output.recovery_train'] = recovery(data_train['rougher.output.concentrate_au'], data_train['rougher.input.feed_au'], data_train['rougher.output.tail_au'])
print('Проверка корректности расчета признака rougher.output.recovery_train в обучающей выборке по сумме: ', data_train['rougher.output.recovery_train'].sum() - data_train['rougher.output.recovery'].sum())
print('Проверка корректности расчета признака rougher.output.recovery_train в обучающей выборке по среднему: ', data_train['rougher.output.recovery_train'].mean() - data_train['rougher.output.recovery'].mean())
```

```python
# расчет MAE расчетного признака обучающей выборки и значения полученного в выборке
mae_rougher_output_recovery_train = mean_absolute_error(data_train['rougher.output.recovery_train'], data_train['rougher.output.recovery'])
print('Средняя абсолютная ошибка(MAE) между полученными значениями обучающей выборки и расчетными значениями: ', mae_rougher_output_recovery_train)
```

## Модель

### Напишем функцию для вычисления итоговой sMAPE.

Необходимо спрогнозировать две величины:
1. эффективность обогащения чернового концентрата rougher.output.recovery;
2. эффективность обогащения финального концентрата final.output.recovery.

```python
# напишем функцию для расчета smape
def smape(y_true, y_pred):
    error = np.abs(y_true - y_pred)
    scale = (np.abs(y_true) + np.abs(y_pred)) / 2
    
    return (error / scale).mean()
```

```python
from sklearn.dummy import DummyRegressor
# модели простого регрессора с кроссвалидацией
model_dr_rougher = DummyRegressor(strategy='median')
model_dr_rougher.fit(data_train_features, data_train_target_rougher)
dr_scores_rougher = cross_val_score(model_dr_rougher, data_train_features, data_train_target_rougher, cv=5, scoring=score)
dr_scores_rougher_mean = sum(dr_scores_rougher) / len(dr_scores_rougher)
print(abs(dr_scores_rougher_mean) * 100)

model_dr_final = DummyRegressor(strategy='median')
model_dr_final.fit(data_train_features, data_train_target_final)
dr_scores_final = cross_val_score(model_dr_final, data_train_features, data_train_target_final, cv=5, scoring=score)
dr_scores_final_mean = sum(dr_scores_final) / len(dr_scores_final)
print(abs(dr_scores_final_mean) * 100)
```

```python
predicted_dr_rougher = model_dr_rougher.predict(data_test_features)
predicted_dr_final = model_dr_final.predict(data_test_features)
```

```python
calculate_final_smape = 0.25 * smape(data_test_target_rougher, predicted_dr_rougher) + 0.75 * smape(data_test_target_final, predicted_dr_final)
print(calculate_final_smape * 100)
```

Простой регрессор с кроссвалидацией и предсказанием медианных значений обучающей выборки показал, что получившиеся выше результаты достаточно близки к результатам простого регрессора, а в каких-то моделях и хуже. Поэтому следует либо дополнительно заняться поиском более результативных моделей, либо применить другие более емкие метрики для оценки качества моделей.

```python
# вариант без кроссвалидации
simple_model_dr_rougher = DummyRegressor(strategy='median')
simple_model_dr_rougher.fit(data_train_features, data_train_target_rougher)
dr_predict_rougher = simple_model_dr_rougher.predict(data_test_features)
dr_smape_rougher = smape(data_test_target_rougher, dr_predict_rougher) * 100
print(dr_smape_rougher)

simple_model_dr_final = DummyRegressor(strategy='median')
simple_model_dr_final.fit(data_train_features, data_train_target_final)
dr_predict_final = simple_model_dr_final.predict(data_test_features)
dr_smape_rougher = smape(data_test_target_final, dr_predict_final) * 100
print(dr_smape_rougher)
```

```python
calculate_final_smape = 0.25 * smape(data_test_target_rougher, dr_predict_rougher) + 0.75 * smape(data_test_target_final, dr_predict_final)
print(calculate_final_smape * 100)
```

Вариант без кроссвалидации также показал, что качество указанных выше моделей в основном хуже без переобучения, но качество модели линейной регрессии отличается незначительно от дамми модели. Поэтому дополнительно к выводам выше может быть еще целесообразно добавить признаки с разбиениемм каждого технологического этапа на подэтапы, т.к. происходит сильно упрощенное и достаточно грубое моделирование на основе ограниченного количества признаков по технологическому процессу.
